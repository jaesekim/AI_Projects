# Feature Engineering

## 변수생성

```
유동비율(%) = 유동자산 / 유동부채 x 100
부채비율 = 총부채 / 자기자본 x 100
비유동장기적합률 = 비유동자산/(자기자본 + 비유동부채)  x 100

ROA = 순이익 / 총자산 x 100
ROE(자기자본순이익률) = 순이익 / 자기자본 x 100
매출액순이익률 = 당기순이익 / 매출액 x 100

총자본회전율 = 매출액 / 총자본
자기자본회전율 = 매출액 / 자기자본

자본집약도 = 총자본 / 종업원 수

- 성장성 비율
매출액증가율 = 당기매출증가액 / 전기매출액 x 100
유동자산증가율 = 당기유동자산증가액 / 전기말유동자산 x 100
순이익증가율 = 당기의 당기순이익 증가액 / 전기 당기순이익 x 100
```

````python
train["유동비율_2017"] = np.round((train["liquidAsset1"] / train["liquidLiabilities1"]) * 100, 2)
test["유동비율_2017"] = np.round((test["liquidAsset1"] / test["liquidLiabilities1"]) * 100, 2)
train["유동비율_2016"] = np.round((train["liquidAsset2"] / train["liquidLiabilities2"]) * 100, 2)
test["유동비율_2016"] = np.round((test["liquidAsset2"] / test["liquidLiabilities2"]) * 100, 2) 
````

````python
train["부채비율_2017"] = np.round((train["debt1"] / train["netAsset1"]) * 100, 2)
test["부채비율_2017"] = np.round((test["debt1"] / test["netAsset1"]) * 100, 2)
train["부채비율_2016"] = np.round((train["debt2"] / train["netAsset2"]) * 100, 2)
test["부채비율_2016"] = np.round((test["debt2"] / test["netAsset2"]) * 100, 2)
````

````python
train["비유동장기적합률_2017"] = np.round((train["nonCAsset1"] / (train["netAsset1"] + train["NCLiabilities1"])) * 100, 2)
test["비유동장기적합률_2017"] = np.round((test["nonCAsset1"] / (test["netAsset1"] + test["NCLiabilities1"])) * 100, 2)

train["비유동장기적합률_2016"] = np.round((train["nonCAsset2"] / (train["netAsset2"] + train["NCLiabilities2"])) * 100, 2)
test["비유동장기적합률_2016"] = np.round((test["nonCAsset2"] / (test["netAsset2"] + test["NCLiabilities2"])) * 100, 2)
````

````python
train["ROA_2017"] = np.round((train["profit1"] / (train["netAsset1"] + train["debt1"])) * 100, 2)
test["ROA_2017"] = np.round((test["profit1"] / (test["netAsset1"] + test["debt1"])) * 100, 2)

train["ROA_2016"] = np.round((train["profit2"] / (train["netAsset2"] + train["debt2"])) * 100, 2)
test["ROA_2016"] = np.round((test["profit2"] / (test["netAsset2"] + test["debt2"])) * 100, 2)
````

````python
train["ROE_2017"] = np.round((train["profit1"] / train["netAsset1"]) * 100, 2)
test["ROE_2017"] = np.round((test["profit1"] / test["netAsset1"]) * 100, 2)

train["ROE_2016"] = np.round((train["profit2"] / train["netAsset2"]) * 100, 2)
test["ROE_2016"] = np.round((test["profit2"] / test["netAsset2"]) * 100, 2)
````

````python
train["매출액순이익률_2017"] = np.round((train["profit1"] / train["revenue1"]) * 100, 2)
test["매출액순이익률_2017"] = np.round((test["profit1"] / test["revenue1"]) * 100, 2)

train["매출액순이익률_2016"] = np.round((train["profit2"] / train["revenue2"]) * 100, 2)
test["매출액순이익률_2016"] = np.round((test["profit2"] / test["revenue2"]) * 100, 2)
````

````python
train["총자본회전율_2017"] = np.round((train["revenue1"] / (train["netAsset1"] + train["debt1"])) * 100, 2)
test["총자본회전율_2017"] = np.round((test["revenue1"] / (test["netAsset1"] + test["debt1"])) * 100, 2)

train["총자본회전율_2016"] = np.round((train["revenue2"] / (train["netAsset2"] + train["debt2"])) * 100, 2)
test["총자본회전율_2016"] = np.round((test["revenue2"] / (test["netAsset2"] + test["debt2"])) * 100, 2)
````

````python
train["자기자본회전율_2017"] = np.round((train["revenue1"] / train["netAsset1"]) * 100, 2)
test["자기자본회전율_2017"] = np.round((test["revenue1"] / test["netAsset1"]) * 100, 2)

train["자기자본회전율_2016"] = np.round((train["revenue2"] / train["netAsset2"]) * 100, 2)
test["자기자본회전율_2016"] = np.round((test["revenue2"] / test["netAsset2"]) * 100, 2)
````

````python
train["매출액증가율"] = np.round(((train["revenue1"] - train["revenue2"]) / train["revenue2"]) * 100, 2)
test["매출액증가율"] = np.round(((test["revenue1"] - test["revenue2"]) / test["revenue2"]) * 100, 2)
````

````python
train["유동자산증가율"] = np.round(((train["liquidAsset1"] - train["liquidAsset2"]) / train["liquidAsset2"]) * 100, 2)
test["유동자산증가율"] = np.round(((test["liquidAsset1"] - test["liquidAsset2"]) / test["liquidAsset2"]) * 100, 2)
````

````python
train["순이익증가율"] = np.round(((train["profit1"] - train["profit2"]) / train["profit2"]) * 100, 2)
test["순이익증가율"] = np.round(((test["profit1"] - test["profit2"]) / test["profit2"]) * 100, 2)
````

# 데이터셋 나누기

````python
label_name = "OC"
label_name
````

````python
train.columns
````

````python
feature_names = [
    'bedCount', '유동비율_2017', '유동비율_2016',
       '부채비율_2017', '부채비율_2016', '비유동장기적합률_2017', '비유동장기적합률_2016', 'ROA_2017',
       'ROA_2016', 'ROE_2017', 'ROE_2016', '매출액순이익률_2017', '매출액순이익률_2016',
       '총자본회전율_2017', '총자본회전율_2016', '자기자본회전율_2017', '자기자본회전율_2016', '매출액증가율',
       '유동자산증가율', '순이익증가율', 'instkind_clinic', 'instkind_dental_clinic',
       'instkind_general_hospital', 'instkind_hospital',
       'instkind_nursing_hospital', 'instkind_traditional_clinic',
       'instkind_traditional_hospital', 'ownerChange_change',
       'ownerChange_same'
]
````

````python
X_train = train[feature_names]
X_train
````

````python
y_train = train[label_name]
y_train.isnull().sum()
````

````python
X_test = test[feature_names]
X_test
````

````python
X_train = X_train.replace([np.inf, -np.inf], np.nan)
X_test = X_test.replace([np.inf, -np.inf], np.nan)
X_train
````

````python
X_train = X_train.replace(np.nan, np.round(X_train.mean(), 2))
X_test = X_test.replace(np.nan, np.round(X_train.mean(), 2))
X_test.isnull().sum().sum()
````

````python
X_train.isnull().sum().sum()
````

````python
input_shape = len(X_train.keys())
input_shape
````

````python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=256, activation='selu', input_shape=[input_shape]),
    tf.keras.layers.Dense(units=128, activation='selu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(units=64, activation='selu'),
    tf.keras.layers.Dense(units=32, activation='selu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(1, activation="sigmoid")
  ])
````

````python
model.summary()
````

````python
# 모델 컴파일
model.compile(optimizer="sgd", 
              loss='binary_crossentropy',
              metrics=['accuracy'])
````

````python
class PrintDot(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('')
        print('.', end='')

early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=100)
````

````python
# 학습하기
EPOCHS = 10

history = model.fit(X_train, y_train,epochs=EPOCHS, verbose=0, validation_split=0.2, callbacks=[early_stop, PrintDot()])
history
````

````python
df_hist  = pd.DataFrame(history.history)
df_hist.tail()
````

````python
# loss, accuracy 값 시각화 
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
df_hist[["loss", "val_loss"]].plot(ax=axes[0])
df_hist[["accuracy", "val_accuracy"]].plot(ax=axes[1])
````

````python
y_pred = model.predict(X_test).flatten()
y_pred
````
