```python
# 라이브러리 불러오기
import pandas as pd
import numpy as np
```

```python
url_imdb = "https://www.imdb.com/chart/top/"
table1= pd.read_html(url_imdb)[0].dropna(axis=1)
table1
table_imdb = table1[["Rank & Title", "IMDb Rating"]]
tmp = table_imdb["Rank & Title"].str.split("\. ")
table_imdb["Rank & Title"] = tmp.str.get(1)
table_imdb
```

```python
page_no = 1
item_list = []
while page_no <= 5:
    url_naver = f"https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=pnt&date=20220524&page={page_no}"
    table2 = pd.read_html(url_naver)[0].dropna(how="all", axis=0).dropna(how="all", axis=1)
    item_list.append(table2)
    page_no += 1
table_naver = pd.concat(item_list)
table_naver = table_naver.reset_index(drop=True)[["영화명", "평점.1"]]
table_naver
```

```python
movie_chart = pd.concat([table_imdb, table_naver], axis=1)
movie_chart.columns = {"imdb", "imdb 평점", "네이버", "네이버 평점"}
movie_chart
```
